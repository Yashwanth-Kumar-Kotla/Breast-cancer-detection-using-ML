```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
BreastCancerData = pd.read_csv("Breast_cancer_dataset.csv")
```

M (Malignant): Indicates a cancerous tumor.
B (Benign): Indicates a non-cancerous tumor.

```python
BreastCancerData.drop('id', axis=1, inplace=True)
BreastCancerData.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
BreastCancerData.tail(20)
```

```python
X = BreastCancerData.drop(columns='diagnosis', axis=1)
Y = BreastCancerData['diagnosis']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,stratify=Y,random_state=2)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=5000, solver='lbfgs')
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, training_data_prediction)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
training_data_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(Y_test, training_data_prediction)
print('Accuracy on training data : ', training_data_accuracy)
```
